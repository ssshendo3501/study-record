```python
# ライブラリインポート
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# データロード
df = sns.load_dataset('iris')
```

```python
# X,y生成
y_col = 'species'
X = df.drop(columns=[y_col])
y = df[y_col]

# hold-out
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# モデル学習
model = LogisticRegression(penalty='none')
model.fit(X_train, y_train)

# 予測
y_pred = model.predict(X_test)
```

## Pythonで式を作る

```python
# Accuracy
correct_predictions = sum(1 for true, pred in zip(y_true, y_pred) if true == pred)
accuracy = correct_predictions / len(y_true)

# Precison, Recall
true_positive = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 1)
false_positive = sum(1 for true, pred in zip(y_true, y_pred) if true == 0 and pred == 1)
false_negative = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 0)

precision = true_positive / (true_positive + false_positive)
recall = true_positive / (true_positive + false_negative)
```

## F1-score

F1-score
- sklearn.metrics.f1_score
    - f1_score(y_true, y_pred)
    - 他の指標と同様に、average引数で'macro'や'micro'など選択可能

```python
from sklearn.metrics import f1_score, recall_score, precision_score

y_true = [1, 1, 0, 1, 0, 0, 1, 1] 
y_pred = [1, 1, 1, 0, 0, 1, 0, 1] 

f1_score(y_true, y_pred)
```

## 多クラスROC

多クラスROC
- 正解ラベルはone-hotの形にする
- macro平均を求めるにはそれぞれのクラス別のfpr, tprを求め、線形補完し平均を取る
- micro平均はonehotの形の正解ラベルと予測した確率リストを１次元配列にしてroc-curve()に入れれば良い

### データ準備

```python
# データロード
df = sns.load_dataset('iris')

# X,y生成
y_col = 'species'
X = df.drop(columns=[y_col])
y = df[y_col]

# hold-out
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# モデル学習
model = LogisticRegression()
model.fit(X_train, y_train)

# 予測
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)
```

### macro平均

```python
# one-hotの形にする
from sklearn.preprocessing import label_binarize
y_test_one_hot = label_binarize(y_test, classes=model.classes_)

# 各クラスのROC
# class0: 'setosa', class1: 'versicolor', class2: 'virginica'
fpr_0, tpr_0, _ = roc_curve(y_test_one_hot[:, 0], y_pred_proba[:, 0])
fpr_1, tpr_1, _ = roc_curve(y_test_one_hot[:, 1], y_pred_proba[:, 1])
fpr_2, tpr_2, _ = roc_curve(y_test_one_hot[:, 2], y_pred_proba[:, 2])

n_classes = len(model.classes_)
fpr = {}
tpr = {}
roc_auc = {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_one_hot[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
# それぞれのクラスのROC
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'class_{i}')
    plt.legend()
plt.xlabel('1-Specifitcity(fpr)')
plt.ylabel('Recall(tpr)')
plt.title('Multi Class ROC Curve')
plt.show()
```

### 線形補完

多クラスROCを描画する



線形補完
- np.interp(x, xp, fp)
    - x: 補完後のx軸の値
    - xp: x軸の値のリスト
    - ｙｐ: y軸の値のリスト
    


np.interpの例

```python
all_x = np.arange(7)
data_x = [0, 2, 5, 6]
data_y = [0, 2, 4, 6]
np.interp(all_x, data_x, data_y)
```

多クラスのROCを線形補完

```python
# 必要なxのリスト
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

# 線形補完
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr = mean_tpr / n_classes
```

```python
fpr['macro'] = all_fpr
tpr['macro'] = mean_tpr
roc_auc['macro'] = (fpr['macro'], tpr['macro'])
```

### micro

```python
fpr['micro'], tpr['micro'], _ = roc_curve(y_test_one_hot.ravel(), y_pred_proba.ravel())
roc_auc['micro'] = auc(fpr['micro'], tpr['micro'])
```

```python
roc_auc
```

### 可視化

```python
# それぞれのクラスとmacro平均とmicro平均のROC
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'class: {i}')
plt.plot(fpr['macro'], tpr['macro'], label='macro')
plt.plot(fpr['micro'], tpr['micro'], label='micro')
plt.legend()
```
