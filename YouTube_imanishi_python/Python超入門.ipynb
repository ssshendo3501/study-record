# zip関数とenumerate関数

image.png

```python
last_names = ['今西', '佐藤', '鈴木', '田中']
first_names = ['航平', '健介', '浩司', '康介']
```

```python
print(last_names[0] + first_names[0] + 'さん')
print(last_names[1] + first_names[1] + 'さん')
print(last_names[2] + first_names[2] + 'さん')
print(last_names[3] + first_names[3] + 'さん')
```

```python
for i in range(len(last_names)):
  print(last_names[i] + first_names[i] + 'さん')
```

```python
range(3)
```

```python
for last_name, first_name in zip(last_names, first_names):
  print(last_name + first_name + 'さん')
```

```python
print('出席番号' ,0,  '番目' + last_names[0] + 'さん')
print('出席番号' ,1,  '番目' + last_names[1] + 'さん')
print('出席番号' ,2,  '番目' + last_names[2] + 'さん')
print('出席番号' ,3,  '番目' + last_names[3] + 'さん')
```

```python
for i in range(len(last_names)):
  print('出席番号' ,i,  '番目' + last_names[i] + 'さん')
```

```python
for i, last_name in enumerate(last_names):
  print('出席番号' ,i,  '番目' + last_name + 'さん')
```

```python
for i, last_name in enumerate(last_names):
  print(f'出席番号{i}番目の{last_name}さん')
```

# break文とcontinue文

```python
for i in range(10):
  print(i)
```

```python
# (0, 10)で繰り返す
# i == 7になるまでprint(i)を繰り返す
for i in range(10):
  if i == 7:
    break
  print(i)
```

```python
# (0, 10)で繰り返す
# i == 7のときをskipして繰り返す
for i in range(10):
  if i == 7:
    continue
  print(i)
```

# クラス

image.png

```python
#クラスは最初のクラス名を大文字で表す

# class (クラス名):
#   def __init__(self, x,y,z):
#

class Person:
  #初期設定
  def __init__(self, name, nationality, age):
    self.name = name
    self.nationality = nationality
    self.age = age

  #クラスで定義される関数は関数ではなくメソッド
  def __call__(self, name):
     print(f'{name}さん、こんにちは。私は{self.name}です。call関数から呼び出されています。')

  def say_hello(self, name):
    print(f'{name}さん、こんにちは。私は{self.name}です。')
```

```python
#今西のインスタンスを作る
imanishi = Person(name='今西', nationality='日本', age=26)
```

```python
imanishi
```

```python
# imanishiのインスタンスの中にはname、nationality、ageの情報が入っている
# self.age = age を呼び出している
# selfはインスタンス自身を意味する引数のこと
imanishi.age
```

```python
imanishi.nationality
```

```python
imanishi.name
```

```python
#マイクのインスタンスを作る
mike = Person(name='マイク', nationality='アメリカ', age=23)
```

```python
mike.age
```

```python
imanishi.age
```

```python
imanishi.say_hello(name='佐藤')
```

```python
mike.say_hello(name='近藤')
```

```python
imanishi.__call__(name = '佐藤')
```

```python
# __cal__は特殊な関数で、インスタンス名()で呼び出すことができる
imanishi(name='佐藤')
```
