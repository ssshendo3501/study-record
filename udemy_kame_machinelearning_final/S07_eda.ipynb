```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
df
```

データの意味

Rank - Ranking of overall sales  
Name - The games name  
Platform - Platform of the games release (i.e. PC,PS4, etc.)  
Year - Year of the game's release  
Genre - Genre of the game  
Publisher - Publisher of the game  
NA_Sales - Sales in North America (in millions)  
EU_Sales - Sales in Europe (in millions)  
JP_Sales - Sales in Japan (in millions)  
Other_Sales - Sales in the rest of the world (in millions)  
Global_Sales - Total worldwide sales.  

```python
# カラム名、データ型、欠損値確認
df.info()
```

```python
# 統計検定量
df.describe()
```

```python
# 各種相関
import seaborn as sns
sns.pairplot(df)
```

```python
# Rankが一番大きいデータは何か？
df[df['Rank']==16600]
```

```python
# 売上が高い順に並び替え
df.sort_values('Global_Sales')
```

```python
# メーカーのごと売上の合計の高い順に並び替え
df.groupby('Publisher').sum().sort_values('Global_Sales', ascending=False)
```

```python
# 可視化
%matplotlib inline
df.groupby('Publisher').sum().sort_values('Global_Sales', ascending=False)[:10].plot.bar(y='Global_Sales')
```

```python
# 西暦のデータの全容確認
df['Year'].value_counts().sort_index()
```

```python
df.sort_values('Year', ascending=False)[:10]
```

```python
# 相関ヒートマップ
sns.heatmap(df.corr(), annot=True)
```

```python
# NA,EU,JP,Otherの合計の売上とGlobal_Salesが一致するか確認
df['NA_EU_JP_Other_Sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']
df['Global_Sales - NA_EU_JP_Other'] = np.abs(df['Global_Sales'] - df['NA_EU_JP_Other_Sales'])

df.sort_values('Global_Sales - NA_EU_JP_Other', ascending=False)
```
